```python
import pandas as pd
import seaborn
import numpy as np

seaborn.set()

# Setting where our data is sitting
data_folder = path.join(
    path.abspath('..'),  # '..' means the directory above this one
    'data')
data_folder
```

```python
os.listdir(data_folder)
```

```python
filename_1 = 'clean_FL_21092018.feather'
filename_2 = 'clean_notifications_19092018.feather'
```

```python
df_FL = pd.read_feather(data_folder+'/'+filename_1)
df_notif = pd.read_feather(data_folder+'/'+filename_2)
```

```python
floc = pd.DataFrame.from_records([df_FL.floc, df_FL.facility_type, df_FL.catalog_profile, df_FL.location],
    columns=('floc','facility_type','catalog_profile','location'),
    index='floc'
)
floc.head()
```

```python
notif = pd.DataFrame.from_records([(df_notif.functional_location, df_notif.damage_code, df_notif.cause_code,
                                      df_notif.object_part, df_notif.location)
                                     for floc, df in df_notif.groupby('functional_location')
                                     ],
    columns=('floc','damage','cause', 'object','location'),
    index='floc'
)
notif.head()
```

```python
pareto = floc.join(notif, how='inner')
```
